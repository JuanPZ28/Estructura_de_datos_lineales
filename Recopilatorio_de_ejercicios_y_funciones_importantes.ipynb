```python
#LISTA CON ARREGLO E INSERCION
class Lista:
    def __init__(self):
        # Lista con capacidad 5, pero solo 4 elementos definidos
        self.data = ["A", "B", "C", "D", None]
        self.capacidad = 5
        self.tamano = 4

    def insert(self, index, element):
        if self.tamano < self.capacidad:
            # Desplazar elementos a la derecha
            for i in range(self.tamano, index, -1):
                self.data[i] = self.data[i - 1]

            # Insertar el nuevo valor
            self.data[index] = element
            self.tamano += 1  # Aumentar tamaño real
        else:
            print("La lista está llena, no se pueden insertar más elementos")

# Prueba
mi_lista = Lista()
print("Lista antes de la insercion", mi_lista.data)
print("Tamaño antes de la inserción:",mi_lista.tamano)

mi_lista.insert(2, "F")  # Insertar "F" en la posición 2

print("\nLista después de la inserción:")
for i in range(mi_lista.capacidad):  # Imprimir hasta la capacidad máxima
    print("Índice" ,i,":",mi_lista.data[i])

print("Lista final:", mi_lista.data)
print("Tamaño después de la inserción:", mi_lista.tamano)
```

```python
#LISTA CIRCULAR
#Ejercicio Lista Circular
class ListaCircular:
    def __init__(self, capacidad):#Funcion para inicializar elementos con el self
      self.capacidad=capacidad
      self.data=[None]*capacidad #Asigna una capacidad fija al arreglo llenandolo con Nones
      self.tamano=0
      self.frente=0

    def insert(self, indice, valor):
      if self.tamano < self.capacidad:#Verificar si aun se pueden insertar elementos dentro de la lista circular

        pos_real = (self.frente + indice) % self.capacidad
        """Busqueda de posicion real:
      Este metodo es utilizado en listas circulares  para calcular una posicon real dentro de ella,
      debidoo a que no siempre el indice '0' sera el frente logico de la lista. El frente de esta lista
      puede estar en cualquier posicion y los elemntos se guardan de manera rotatoria. por ello el self.frente
      indica el inicio de la lista logica y el indice representa el desplazamiento que se inicia desde ese frente
      por ello se suman estos dos valores y para evitar que se salga del arreglo, ahi entra en juego el modulo '%'"""
        for i in range(self.tamano, indice, -1):
          self.data[(self.frente + i) % self.capacidad] = self.data[(self.frente + i - 1) % self.capacidad]
        self.data[pos_real] = valor
        self.tamano += 1

      else:
            # Buffer lleno: sobrescribir el elemento más antiguo
            if indice == 0:
                # Si se inserta al inicio, simplemente sobrescribe el frente actual
                self.data[self.frente] = valor
                self.frente = (self.frente + 1) % self.capacidad  # Avanzamos el frente

    def rotar(self, k=1):
      self.frente = (self.frente + k) % self.capacidad

    def eliminar(self, indice):
      if 0 <= indice < self.tamano:
        pos_real = (self.frente + indice) % self.capacidad
        for i in range(pos_real, self.tamano - 1):
          self.data[i] = self.data[i + 1]
        self.data[self.tamano - 1] = None
        self.tamano -= 1

    def is_empty(self):
      return self.tamano == 0

    def size(self):
      return self.tamano

    def esta_lleno(self):
      return self.tamano == self.capacidad

    def get_item(self, indice):
      if 0 <= indice < self.tamano:
        return self.data[(self.frente + indice) % self.capacidad]
```

```python
#IMPLEMENTACION PILA ENLAZADA
#Una pila con lista enlazada
from graphviz import Digraph
class Nodo:#Clase nodo que sirve para la asignacion del valor actual y la asignacion del valor siguiente
    def __init__(self, valor):
        self.valor = valor
        self.siguiente = None

class PilaEnlazada:
    def __init__(self):
        self.cima = None#Establece un valor en la cima

    def push(self, valor):#Empieza a realizar la operacion de insercion de datos como si de una pila fuera
    # Cuando entra un nuevo valor, la cima pasa a ser el siguiente valor y el nuevo valor pasa a ser la cima
        nuevo_nodo = Nodo(valor)
        nuevo_nodo.siguiente = self.cima
        self.cima = nuevo_nodo

    def pop(self): #Desencolamiento, comprueba si primero hay algo que desencolar
        if self.is_empty():
            return None
        valor = self.cima.valor# Toma el valor que se encuentra a la cabeza y lo saca de la pila,
        #y convierte el valor abajo del sacado en la nueva cima
        self.cima = self.cima.siguiente
        return valor#retorna el valor sacado de la fila

    def is_empty(self):
        return self.cima is None

    def visualizar_pila(self):
      dot = Digraph()

      actual = self.cima
      indice = 0

      while actual is not None:
        dot.node(f'n{indice}', str(actual.valor))
        if actual.siguiente is not None:
            dot.edge(f'n{indice}', f'n{indice+1}')
        actual = actual.siguiente
        indice += 1
        display(dot)

Pila=PilaEnlazada()
Pila.push(1)
Pila.push(2)
Pila.push(3)
Pila.push(4)
Pila.push(5)
Pila.push(6)
Pila.visualizar_pila()
```

```python
#IMPLEMENTACION Y USO DE LA PILA CON LISTAS
import time
import sys
class Pila:#uso de pila con listas

  def __init__(self):
    self.items = []

  def is_empty(self):
    return len(self.items) == 0

  def push(self, item):
    self.items.append(item)

  def pop(self):
    if self.is_empty():
      raise IndexError("La pila está vacía")
    return self.items.pop()

  def peek(self):
    if self.is_empty():
      raise IndexError("La pila está vacía")
    return self.items[-1]

  def size(self):
    return len(self.items)
#4-----------------------------------------------------------------------
  def imprimir(self):
    if self.is_empty():
      print("La pila está vacía")
    else:
      print("Elementos de la pila:")
      for item in reversed(self.items):
        print(item)




if __name__ == "__main__":
  pila = Pila()
  pila.push(1)
  pila.push(2)
  pila.push(3)
  pila.push(4)
  pila.push(5)
  print(pila.peek())
  pila.pop()
  pila.pop()
  print(pila.peek())
  print(pila.is_empty())
 #4------------------------------------------------------------------------
  pila.imprimir()
  print(pila.peek())
```

```python
#PILA EN LA SOLUCION DE PROBLEMAS RECURSIVOS, FACTORIAL
#Uso de pila en la solucion de una recursividad

def factorial_con_pila(n):
  if n<0:
    raise ValueError("El factorial no está definido para números negativos")
  pila = []
  #LLenado de la pila con los elementos del factorial
  for i in range(1, n+1):
    pila.append(i)
  resultado = 1
  while pila:
    #multiplicador d elos elementos
    resultado *= pila.pop()
  return resultado

print(factorial_con_pila(5))
```

```python
class Cola:
  def __init__(self):#Inicializa items
    self.items = []

  def encolar(self, item): # Encola los datos en la cola
    self.items.append(item)
    print("Encolado", item)

  def desencolar(self): # Desencola los datos en la cola
    if not self.esta_vacia(): #Revisa si la cola no está vacía
      item = self.items.pop(0)
      print("Desencolado", item)
    else:
      print("La cola está vacía")

  def esta_vacia(self): #Revisa si la cola está vacía
    return len(self.items) == 0

  def frente(self): #Obtiene el frente de la cola
    if not self.esta_vacia(): #Revisa si la cola no está vacía
      return self.items[0] #Devuelve el elemento en la posición 0, osea, el inicio de la cola
    else:
      print("La cola está vacía")
      return None

  def tamano(self): #Obtiene el tamaño de la cola
      return len(self.items) #Deuvelve la longitud de la lista

#Ejemplo de uso
if __name__ == "__main__":
  cola= Cola()
  cola.encolar("A")
  cola.encolar("B")
  cola.encolar("C")
  cola.desencolar()
  print(cola.frente())
  print(cola.tamano())
  print(cola.esta_vacia())
```

```python
#IMPLEMENTACION DE COLA CIRCULAR
class ColaCircular:
  def __init__(self, capacidad): #Crea una cola circular, que detecte el frente, el último valor, la capacidad y el tamaño de la cola
    self.items = [None] * capacidad
    self.capacidad = capacidad
    self.primero = 0
    self.ultimo = -1
    self.tamano = 0

  def encolar(self, item): #Encola los datos en la cola
    if self.tamano == self.capacidad: #Revisa si el tamaño es igual a la capacidad de la cola
      print("La cola está llena")
    elif self.esta_vacia(): #Revisa si la cola está vacía y ejecuta este código
      self.primero = 0 #El frente será 0
      self.ultimo = 0 #El último dato de la cola está en posición 0
      self.items[self.ultimo] = item #Agrega el item a la cola
      self.tamano += 1 #Agrega 1 al tamaño
      print("Encolado", item)
    else: #Si ya hay valores en la cola, ejecuta
      self.ultimo = (self.ultimo + 1) % self.capacidad #El último valor será el residuo del anterior último valor + 1 con la capacidad de la cola
      self.items[self.ultimo] = item #guarda en el último valor el item
      self.tamano += 1 #Agrega 1 al tamaño
      print("Encolado", item)

  def desencolar(self): #Desnecola los datos de la lista
    if self.esta_vacia(): #Revisa si la cola está vacía
      print("La cola está vacía")
    elif self.primero == self.ultimo: #Revisa si solo hay un dato en la cola
      item = self.items[self.primero] #Obtiene el item de la cola
      self.items.pop(0) #Elimina el item
      self.primero = 0 #El frente de la cola es 0
      self.ultimo = -1 #El último valor será -1, debido a que ya no hay datos
      self.tamano -= 1 #Resta 1 al tamaño
      print("Desencolado", item)
    elif self.primero < self.capacidad - 1: #Revisa si el frente es menor a la capacidad -1
      item = self.items[self.primero] #Obtiene el item de la cola
      self.items.pop(0) #Elimina el item
      self.primero = 0 #El frente de la cola es 0
      self.tamano -= 1 #Resta el tamaño de la cola
      print("Desencolado", item)

  def esta_vacia(self): #Revisa si la cola está vacía
    return self.tamano == 0

  def frente(self): #Revisa el frente de la lista
    if not self.esta_vacia():
      return self.items[self.primero]
    else:
      print("La cola está vacía")
      return None

  def tamaño(self): #Obtiene el tamaño de la lista
    return self.tamano

  def visualize(self): #Visualiza la lista
    print(self.items)


#Ejemplo de uso
if __name__ == "__main__":
  cola= ColaCircular(15)
  for i in range(15):
    cola.encolar(i+1)
  cola.visualize()
  cola.desencolar()
  cola.visualize()
  print(cola.frente())
  print(cola.tamaño())
  print(cola.esta_vacia())
  cola.desencolar()
  cola.visualize()
  print(cola.frente())
  print(cola.tamaño())
  print(cola.esta_vacia())
```

```python
#COLA DE PRIORIDAD

"---------------------Ejercicio 1------------------------------------"


import heapq

class PriorityQueue:
  def __init__(self):
    self.heap = []
    self.count = 0

  def push(self, item, priority):

    heapq.heappush(self.heap, (priority, self.count, item))
    self.count += 1

  def pop(self):
    if self.heap:
      priority, count, item = heapq.heappop(self.heap)
      return item
    raise IndexError("pop from an empty priority queue")

  def is_empty(self):
    return len(self.heap) == 0

pq= PriorityQueue()
print("Lista de pacientes:")
pq.push("Daniel Oviedo", 1)
pq.push("Camilo Diaz", 3)
pq.push("Juan David", 2)

print("Pacientes atentidos por orden de prioridad:")
while not pq.is_empty():
  print(pq.pop())


```

```python
 #BICOLA: S EPUEDEN INSERTAR Y ELIMINAR ELEMENTOS DESDE AMBOS EXTREMOS
  "---------------------Ejercicio 2------------------------------------"


"""
Implementación de una Bicola (Double-Ended Queue) en Python
utilizando el módulo collections.deque.

Este programa permite al usuario interactuar con una bicola
a través de un menú para realizar operaciones de inserción
y eliminación en ambos extremos.
"""

from collections import deque

class SistemaBicola:
    """
    Sistema para gestionar una bicola (deque) con operaciones
    de inserción y eliminación en ambos extremos.
    """

    def __init__(self):
        """
        Inicializa una nueva bicola vacía utilizando collections.deque.
        """
        self.bicola = deque()

    def insertar_final(self, elemento):
        """
        Inserta un elemento al final de la bicola.

        Args:
            elemento: El elemento a insertar al final.
        """
        self.bicola.append(elemento)
        print(f"Elemento '{elemento}' añadido al final de la bicola.")

    def insertar_inicio(self, elemento):
        """
        Inserta un elemento al inicio de la bicola.

        Args:
            elemento: El elemento a insertar al inicio.
        """
        self.bicola.appendleft(elemento)
        print(f"Elemento '{elemento}' añadido al inicio de la bicola.")

    def eliminar_final(self):
        """
        Elimina y devuelve el elemento del final de la bicola.

        Returns:
            El elemento eliminado del final o None si la bicola está vacía.
        """
        if not self.esta_vacia():
            elemento = self.bicola.pop()
            print(f"Elemento '{elemento}' eliminado del final de la bicola.")
            return elemento
        else:
            print("No se puede eliminar: la bicola está vacía.")
            return None

    def eliminar_inicio(self):
        """
        Elimina y devuelve el elemento del inicio de la bicola.

        Returns:
            El elemento eliminado del inicio o None si la bicola está vacía.
        """
        if not self.esta_vacia():
            elemento = self.bicola.popleft()
            print(f"Elemento '{elemento}' eliminado del inicio de la bicola.")
            return elemento
        else:
            print("No se puede eliminar: la bicola está vacía.")
            return None

    def esta_vacia(self):
        """
        Verifica si la bicola está vacía.

        Returns:
            bool: True si la bicola está vacía, False en caso contrario.
        """
        return len(self.bicola) == 0

    def mostrar_estado(self):
        """
        Muestra el estado actual de la bicola.
        """
        print("\n=== Estado Actual de la Bicola ===")
        if self.esta_vacia():
            print("La bicola está vacía.")
        else:
            print(f"Contenido de la bicola: {list(self.bicola)}")
            print(f"Elemento al inicio: {self.bicola[0]}")
            print(f"Elemento al final: {self.bicola[-1]}")
            print(f"Tamaño de la bicola: {len(self.bicola)} elementos")
        print("================================\n")


def mostrar_menu():
    """
    Muestra el menú de opciones disponibles para el usuario.
    """
    print("\n==== SISTEMA DE BICOLA (DEQUE) ====")
    print("1. Insertar elemento al final")
    print("2. Insertar elemento al inicio")
    print("3. Eliminar elemento del final")
    print("4. Eliminar elemento del inicio")
    print("5. Mostrar estado de la bicola")
    print("0. Salir")
    print("==================================")
    return input("Seleccione una opción: ")


def ejecutar_sistema_bicola():
    """
    Función principal que ejecuta el sistema de bicola interactivo.
    """
    print("Bienvenido al Sistema de Bicola (Double-Ended Queue)")
    print("---------------------------------------------------")
    print("Este sistema permite operar con una estructura de datos")
    print("que admite inserciones y eliminaciones en ambos extremos.")

    sistema = SistemaBicola()

    while True:
        opcion = mostrar_menu()

        if opcion == '1':
            elemento = input("Ingrese el elemento a insertar al final: ")
            sistema.insertar_final(elemento)
            sistema.mostrar_estado()

        elif opcion == '2':
            elemento = input("Ingrese el elemento a insertar al inicio: ")
            sistema.insertar_inicio(elemento)
            sistema.mostrar_estado()

        elif opcion == '3':
            sistema.eliminar_final()
            sistema.mostrar_estado()

        elif opcion == '4':
            sistema.eliminar_inicio()
            sistema.mostrar_estado()

        elif opcion == '5':
            sistema.mostrar_estado()

        elif opcion == '0':
            print("\nGracias por utilizar el Sistema de Bicola.")
            print("Finalizando programa...")
            break

        else:
            print("\nOpción no válida. Por favor, intente nuevamente.")


def demostrar_operaciones_bicola():
    """
    Función que demuestra las operaciones básicas de una bicola
    utilizando directamente collections.deque.
    """
    print("\n=== DEMOSTRACIÓN DE OPERACIONES BÁSICAS DE BICOLA ===")

    # Crear una bicola
    d = deque()
    print("Bicola creada:", list(d))

    # Insertar elementos en ambos extremos
    print("\n# Insertar elementos en ambos extremos")
    d.append("Elemento final")
    print("Después de append('Elemento final'):", list(d))

    d.appendleft("Elemento inicio")
    print("Después de appendleft('Elemento inicio'):", list(d))

    d.append("Otro elemento final")
    print("Después de append('Otro elemento final'):", list(d))

    # Eliminación de elementos
    print("\n# Eliminación de elementos")
    elem_inicio = d.popleft()
    print(f"Eliminado del inicio: {elem_inicio}")
    print("Bicola actual:", list(d))

    elem_final = d.pop()
    print(f"Eliminado del final: {elem_final}")
    print("Bicola final:", list(d))

    print("\n=== FIN DE LA DEMOSTRACIÓN ===")
    input("\nPresione Enter para continuar al menú interactivo...")


# Ejecución del programa
if __name__ == "__main__":
    # Primero mostrar una demostración de las operaciones básicas
    demostrar_operaciones_bicola()

    # Luego ejecutar el sistema interactivo
    ejecutar_sistema_bicola()

```

```python
#USO DE COLA CIRCULAR, EVITA ESPACIOS EN VACIO

#-------------------Ejercicio 3---------------------------------------------------------------------
class ColaCircular:
    """
    Implementación de una estructura de datos de Cola Circular.

    Una cola circular es una variante de cola que utiliza un arreglo de tamaño fijo
    y dos índices (frente y final) para gestionar las inserciones y extracciones.
    Cuando se llega al final del arreglo, se "envuelve" al inicio si hay espacio
    disponible, solucionando el problema de espacios vacíos que se generan en una
    cola lineal estática.
    """

    def __init__(self, tamano):
        """
        Inicializa una nueva cola circular con el tamaño especificado.

        Args:
            tamano (int): El tamaño máximo de la cola circular.
        """
        self.tamano = tamano
        self.cola = [None] * tamano  # Arreglo de elementos inicializado con None
        self.frente = -1  # Índice que apunta al frente de la cola (primer elemento)
        self.final = -1   # Índice que apunta al final de la cola (último elemento)

    def is_empty(self):
        """
        Verifica si la cola está vacía.

        Returns:
            bool: True si la cola está vacía, False en caso contrario.
        """
        return self.frente == -1

    def is_full(self):
        """
        Verifica si la cola está llena.

        Returns:
            bool: True si la cola está llena, False en caso contrario.
        """
        # La cola está llena si el siguiente elemento después del final
        # coincide con el frente, aplicando el operador módulo para "envolver"
        return (self.final + 1) % self.tamano == self.frente

    def enqueue(self, elemento):
        """
        Añade un elemento al final de la cola.

        Args:
            elemento: El elemento a añadir a la cola.

        Raises:
            Exception: Si la cola está llena.
        """
        if self.is_full():
            raise Exception("La cola está llena")

        # Si la cola está vacía, inicializar frente
        if self.is_empty():
            self.frente = 0

        # Actualizar el índice final usando el operador módulo
        self.final = (self.final + 1) % self.tamano
        self.cola[self.final] = elemento

    def dequeue(self):
        """
        Extrae el elemento del frente de la cola.

        Returns:
            El elemento extraído de la cola.

        Raises:
            Exception: Si la cola está vacía.
        """
        if self.is_empty():
            raise Exception("La cola está vacía")

        # Obtener el elemento a extraer
        elemento = self.cola[self.frente]

        # Si es el único elemento, reiniciar la cola
        if self.frente == self.final:
            self.frente = -1
            self.final = -1
        else:
            # Avanzar el índice frente usando el operador módulo
            self.frente = (self.frente + 1) % self.tamano

        return elemento

    def __str__(self):
        """
        Representa la cola como una cadena de texto.

        Returns:
            str: Una representación en texto de los elementos de la cola.
        """
        if self.is_empty():
            return "Cola vacía"

        elementos = []
        i = self.frente

        while True:
            elementos.append(str(self.cola[i]))
            if i == self.final:
                break
            i = (i + 1) % self.tamano

        return "[" + ", ".join(elementos) + "]"

    def estado_interno(self):
        """
        Muestra el estado interno de la cola para propósitos de depuración.

        Returns:
            str: Una representación detallada del estado interno de la cola.
        """
        return f"Tamaño: {self.tamano}, Frente: {self.frente}, Final: {self.final}, Cola: {self.cola}"


# Función para mostrar el estado de la cola
def mostrar_estado(cola, accion):
    """
    Muestra el estado actual de la cola después de una acción.

    Args:
        cola (ColaCircular): La cola circular a mostrar.
        accion (str): Descripción de la acción realizada.
    """
    print(f"=== {accion} ===")
    print(f"Estado: {cola.estado_interno()}")
    print(f"Cola: {cola}")
    print(f"¿Está vacía? {cola.is_empty()}")
    print(f"¿Está llena? {cola.is_full()}")
    print()


# Pruebas de la Cola Circular
def pruebas_cola_circular():
    """
    Realiza pruebas exhaustivas de la implementación de la cola circular.
    Muestra la evolución del estado interno de la cola después de cada operación.
    """
    print("PRUEBAS DE COLA CIRCULAR")
    print("========================\n")

    # Crear una cola circular de tamaño 5
    cola = ColaCircular(5)
    mostrar_estado(cola, "Inicialización")

    # Prueba 1: Encolar elementos
    print("Prueba 1: Encolar elementos")
    for i in range(1, 6):
        cola.enqueue(i * 10)
        mostrar_estado(cola, f"Encolar {i * 10}")

    # Prueba 2: Intento de encolar en una cola llena
    print("Prueba 2: Intento de encolar en una cola llena")
    try:
        cola.enqueue(60)
    except Exception as e:
        print(f"Error capturado: {e}")
    print()

    # Prueba 3: Desencolar elementos
    print("Prueba 3: Desencolar elementos")
    for _ in range(3):
        elemento = cola.dequeue()
        mostrar_estado(cola, f"Desencolar {elemento}")

    # Prueba 4: Encolar después de desencolar (prueba de circularidad)
    print("Prueba 4: Encolar después de desencolar")
    for i in range(6, 9):
        cola.enqueue(i * 10)
        mostrar_estado(cola, f"Encolar {i * 10}")

    # Prueba 5: Vaciar la cola
    print("Prueba 5: Vaciar la cola")
    while not cola.is_empty():
        elemento = cola.dequeue()
        mostrar_estado(cola, f"Desencolar {elemento}")

    # Prueba 6: Intento de desencolar de una cola vacía
    print("Prueba 6: Intento de desencolar de una cola vacía")
    try:
        cola.dequeue()
    except Exception as e:
        print(f"Error capturado: {e}")


# Ejecutar las pruebas
if __name__ == "__main__":
    pruebas_cola_circular()
```

```python
#MERGE SORT
def merge_sort(arr): #Se define la función para ordenamiento por combinación
  if len(arr) <= 1: #Casos bases de los arreglos. Si sus tamaños son 1 o 0, entonces devuelva el arreglo
    return arr
  mid = len(arr) // 2 #Obtiene el punto medio del arreglo
  left = merge_sort(arr[:mid]) #Obtiene la parte de la izquierda del arreglo, desde 0 hasta la mitad
  right = merge_sort(arr[mid:]) #Obtiene la parte de la derecha del arreglo, desde la mitad hasta el final
  return merge(left, right) #Llama la función "merge" para fusionar la izquierda y la derecha obtenidas

def merge(left, right): #Se define la función para la fusión de los arreglos
  result = []
  i = j = 0
  while i < len(left) and j < len(right): #Mientras los contadores no superen el tamaño de los arreglos
    if left[i] <= right[j]: #Revisa si el arreglo en posición i de la izquierda es mayor o igual a la posición i del arreglo de la derecha
      result.append(left[i]) #Agrega el valor de la izquierda en resultado
      i += 1 #Suma al contador
    else:
      result.append(right[j]) #Si es menor, entonces añade el de la derecha en resultado
      j += 1 #Suma al contador
  while i < len(left): #Mientras que el contador no supere el tamaño del arreglo de la izquierda
    result.append(left[i]) #Agrega a resultado el valor de la izquierda contenido
    i += 1 #Suma al contador
  while j < len(right): #Mientras que el contador no supere el tamaño del arreglo de la derecha
    result.append(right[j]) #Agrega a resultado el valor de la derecha contenido
    j += 1 #Suma al contador
  return result #Devuelve el arreglo

#Ejemplos de uso
arr = [5, 12, 20, 3, 8, 15, 25]
sorted_arr = merge_sort(arr)
print(sorted_arr)

arr = [20, 30, 10, 43, 21, 1, 9, 3]
sorted_arr = merge_sort(arr)
print(sorted_arr)
```

```python
# COMPARATIVA DE USO DE MEMORIA COMPARADO CASOS DE MERGE SORT
import sys

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    while i < len(left):
        result.append(left[i])
        i += 1
    while j < len(right):
        result.append(right[j])
        j += 1
    return result

def measure_memory_usage(arr):
    print(f"Memoria inicial: {sys.getsizeof(arr)} bytes")
    sorted_arr = merge_sort(arr)
    print(f"Memoria final: {sys.getsizeof(sorted_arr)} bytes")
    return sorted_arr

# Ejemplos de uso
arr1 = [5, 12, 20, 3, 8, 15, 25]
print("Arreglo 1:")
sorted_arr1 = measure_memory_usage(arr1)
print("Arreglo ordenado:", sorted_arr1)

arr2 = [20, 30, 10, 43, 21, 1, 9, 3, 23, 12, 67, 89, 90, 21, 54, 67]
print("\nArreglo 2:")
sorted_arr2 = measure_memory_usage(arr2)
print("Arreglo ordenado:", sorted_arr2)
```

```python
#DEIFNICION DEL ARBOL, CON SUS RECORRIDOS
class Nodo:  # Se crea una clase llamada Nodo para representar cada elemento del árbol binario
    def __init__(self, valor):
        self.valor = valor  # Se asigna el valor del nodo
        self.izquierda = None  # Se establece el nodo hijo izquierdo como None (vacío por defecto)
        self.derecha = None  # Se establece el nodo hijo derecho como None (vacío por defecto)

    def inorden(self):
        if self.izquierda:  # Si hay un nodo hijo izquierdo, se llama recursivamente al método inorden
            self.izquierda.inorden()
        print(self.valor, end=' ')  # Se imprime el valor del nodo actual
        if self.derecha:  # Si hay un nodo hijo derecho, se llama recursivamente al método inorden
            self.derecha.inorden()

    def preorden(self):
        print(self.valor, end=' ')  # Se imprime el valor del nodo actual.
        if self.izquierda:  # Si hay un nodo hijo izquierdo, se llama recursivamente al método preorden
            self.izquierda.preorden()
        if self.derecha:  # Si hay un nodo hijo derecho, se llama recursivamente al método preorden
            self.derecha.preorden()

    def postorden(self):
        if self.izquierda:  # Si hay un nodo hijo izquierdo, se llama recursivamente al método postorden
            self.izquierda.postorden()
        if self.derecha:  # Si hay un nodo hijo derecho, se llama recursivamente al método postorden
            self.derecha.postorden()
        print(self.valor, end=' ')  # Se imprime el valor del nodo actual después de visitar a sus hijos

    def insercion(self, valor):
        if valor < self.valor:  # Si el valor es menor, va a la izquierda
            if self.izquierda is None:
                self.izquierda = Nodo(valor)  # Inserta el nodo si está vacío
            else:
                self.izquierda.insercion(valor)  # Sigue buscando dónde insertarlo
        else:  # Si el valor es mayor o igual, va a la derecha
            if self.derecha is None:
                self.derecha = Nodo(valor)  # Inserta el nodo si está vacío
            else:
                self.derecha.insercion(valor)  # Sigue buscando dónde insertarlo

    def busqueda(self, valor):
        if valor == self.valor:
            return True  # Se encontró el valor en el nodo actual
        elif valor < self.valor:
            if self.izquierda is None:
                return False  # No se encontró en la rama izquierda
            else:
                return self.izquierda.busqueda(valor)  # Busca recursivamente a la izquierda
        else:
            if self.derecha is None:
                return False  # No se encontró en la rama derecha
            else:
                return self.derecha.busqueda(valor)  # Busca recursivamente a la derecha

    def altura(self):
        if self is None:
            return -1  # Un árbol vacío tiene altura -1
        else:
            altura_izquierda = self.izquierda.altura() if self.izquierda else -1  # Calcula la altura del subárbol izquierdo si existe, en caso contrario se considera -1
            altura_derecha = self.derecha.altura() if self.derecha else -1 # Calcula la altura del subárbol derecho si existe, en caso contrario se considera -1
            return 1 + max(altura_izquierda, altura_derecha)  # Retorna 1 más la altura mayor entre ambos subárboles.
            # El +1 representa el nivel actual del nodo.

    def recorridopornivel(self):
        cola = deque()
        cola.append(self)
        while cola:
            nodo_actual = cola.popleft()
            print(nodo_actual.valor, end=' ')
            if nodo_actual.izquierda:
                cola.append(nodo_actual.izquierda)
            if nodo_actual.derecha:
                cola.append(nodo_actual.derecha)

# Construcción del árbol realizado en el ejercicio 1
raiz = Nodo(5)
# Crear los nodos según la estructura de la imagen
raiz.izquierda = Nodo(4)
raiz.derecha = Nodo(8)
raiz.izquierda.izquierda = Nodo(2)
raiz.derecha.izquierda = Nodo(7)
raiz.derecha.derecha = Nodo(9)
raiz.derecha.izquierda.izquierda = Nodo(6)  # 6 es hijo izquierdo de 7

# Se realizan los recorridos en el árbol y se muestran los resultados
print("Recorrido Inorden:")
raiz.inorden()
print("\nRecorrido Preorden:")
raiz.preorden()
print("\nRecorrido Postorden:")
raiz.postorden()
print()
#Se inserta un nuevo valor
raiz.insercion(19)
print("Recorrido Inorden:")
raiz.inorden()
print("\nRecorrido Preorden:")
raiz.preorden()
print("\nRecorrido Postorden:")
raiz.postorden()
print()
#Se busca ese valor
print(raiz.busqueda(19))  # Se imprime el resultado de la búsqueda
print()
print("Altura del árbol:", raiz.altura())  # Se imprime la altura del árbol
```

```python
# Ejercicio 5: Comprobar si es un bst

"""
DEFINICIÓN DE LA ESTRUCTURA DEL NODO DEL ÁRBOL

Cada nodo del árbol binario contiene:
- valor: el dato almacenado en el nodo (en este caso, un número entero)
- izquierdo: referencia al hijo izquierdo (None si no existe)
- derecho: referencia al hijo derecho (None si no existe)
"""
class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.izquierdo = None
        self.derecho = None
"""
FUNCIÓN PRINCIPAL PARA VERIFICAR BST

Esta función recursiva verifica si un subárbol cumple con las propiedades BST:
1. Todos los valores en el subárbol izquierdo deben ser menores que el valor del nodo actual
2. Todos los valores en el subárbol derecho deben ser mayores que el valor del nodo actual

Parámetros:
- nodo: el nodo actual que estamos evaluando
- minimo: el valor mínimo permitido para este nodo (inicialmente -infinito)
- maximo: el valor máximo permitido para este nodo (inicialmente +infinito)
"""
def verificar_bst(nodo, minimo=float('-inf'), maximo=float('inf')):
    """
    Verifica si un árbol binario es un BST válido usando el enfoque de rangos.

    Args:
        nodo: El nodo actual que se está evaluando
        minimo: El valor mínimo permitido para este nodo
        maximo: El valor máximo permitido para este nodo

    Returns:
        bool: True si el subárbol es BST válido, False en caso contrario
    """
    # Un árbol vacío es considerado BST válido
    if nodo is None:
        return True

    # Verificar si el valor del nodo está dentro del rango permitido
    if nodo.valor <= minimo or nodo.valor >= maximo:
        return False

    # Verificar recursivamente los subárboles
    # Para el subárbol izquierdo, el valor máximo es el valor del nodo actual
    # Para el subárbol derecho, el valor mínimo es el valor del nodo actual
    return (verificar_bst(nodo.izquierdo, minimo, nodo.valor) and
            verificar_bst(nodo.derecho, nodo.valor, maximo))

def es_arbol_bst(raiz):
    """
    Función wrapper para iniciar la verificación con rangos infinitos.
    """
    return verificar_bst(raiz)

# Pruebas del algoritmo
if __name__ == "__main__":
    # Creamos el árbol con la estructura correcta
    #        5
    #      /   \
    #     4     8
    #    /     /  \
    #   2     7    9
    #        /
    #       6
    arbol_bst = Nodo(5)
    arbol_bst.izquierda = Nodo(4)
    arbol_bst.derecha = Nodo(8)
    arbol_bst.izquierda.izquierda = Nodo(2)
    arbol_bst.derecha.izquierda = Nodo(7)
    arbol_bst.derecha.derecha = Nodo(9)
    arbol_bst.derecha.izquierda.izquierda = Nodo(6)  # 6 es hijo izquierdo de 7


    # Creamos un árbol que NO es BST
    #        50
    #      /    \
    #    30      70
    #   /  \    /  \
    # 20   60 40    80
    arbol_no_bst = Nodo(50)
    arbol_no_bst.izquierdo = Nodo(30)
    arbol_no_bst.derecho = Nodo(70)
    arbol_no_bst.izquierdo.izquierdo = Nodo(20)
    arbol_no_bst.izquierdo.derecho = Nodo(60)  # Violación: 60 > 50
    arbol_no_bst.derecho.izquierdo = Nodo(40)  # Violación: 40 < 50
    arbol_no_bst.derecho.derecho = Nodo(80)

    # Verificamos los árboles
    print("El primer árbol ES un BST válido" if es_arbol_bst(arbol_bst)
          else "El primer árbol NO es un BST válido")

    print("El segundo árbol ES un BST válido" if es_arbol_bst(arbol_no_bst)
          else "El segundo árbol NO es un BST válido")
```

```python
# Ejercicio 6: Recorrido por niveles del arbol binario BFS
"""
DEFINICIÓN DE LA ESTRUCTURA DEL NODO DEL ÁRBOL

Cada nodo del árbol binario contiene:
- valor: el dato almacenado en el nodo (puede ser cualquier tipo de dato)
- izquierdo: referencia al hijo izquierdo (None si no existe)
- derecho: referencia al hijo derecho (None si no existe)
"""

class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.izquierda = None  # Corregido: antes era "izquierdo"
        self.derecha = None    # Corregido: antes era "derecho"
"""
FUNCIÓN PARA RECORRIDO POR NIVELES (BFS)

Este algoritmo implementa el recorrido por niveles de un árbol binario.
Visita los nodos nivel por nivel, comenzando desde la raíz, luego sus hijos, luego los hijos
de sus hijos, y así sucesivamente.

Características principales:
- Utiliza una cola (FIFO) para gestionar el orden de visita
- Procesa todos los nodos de un nivel antes de pasar al siguiente
- Es útil para visualizar la estructura completa del árbol

Parámetros:
- raiz: el nodo raíz del árbol binario a recorrer
"""
def recorrido_por_niveles(raiz):
    """
    Realiza un recorrido por niveles (BFS) de un árbol binario.

    Args:
        raiz: Nodo raíz del árbol binario

    Returns:
        None (imprime los valores de los nodos en orden BFS)
    """
    if raiz is None:
        return

    # Creamos una cola para el BFS
    from collections import deque
    cola = deque()
    cola.append(raiz)

    print("Recorrido por niveles (BFS):")
    while cola:

        # Procesamos el nodo actual
        nodo_actual = cola.popleft()
        print(nodo_actual.valor, end=" ")

        # Agregamos los hijos a la cola si existen
        if nodo_actual.izquierda:
            cola.append(nodo_actual.izquierda)
        if nodo_actual.derecha:
            cola.append(nodo_actual.derecha)
    print()  # Salto de línea al final

# Ejemplo de uso
if __name__ == "__main__":
    # Construimos el árbol binario con la estructura correcta
    #       5
    #      /  \
    #     4    8
    #    /    /  \
    #   2    7    9
    #       /
    #      6
    raiz = Nodo(5)
    raiz.izquierda = Nodo(4)
    raiz.derecha = Nodo(8)
    raiz.izquierda.izquierda = Nodo(2)
    raiz.derecha.izquierda = Nodo(7)
    raiz.derecha.derecha = Nodo(9)
    raiz.derecha.izquierda.izquierda = Nodo(6)  # 6 es hijo izquierdo de 7

    # Realizamos el recorrido por niveles
    recorrido_por_niveles(raiz)
```

```python
#LISTA DOBLEMENTA ENLAZADA
from IPython import get_ipython
from IPython.display import display
from graphviz import Digraph

class DoublyNode:
  def __init__(self, data):
      self.data = data
      self.prev = None
      self.next = None


class DoublyLinkedList:
  def __init__(self):
      self.head = None
      self.tail = None

  def insert_at_beginning(self, data):
      new_node = DoublyNode(data)
      if self.head is None:
          self.head = new_node
          self.tail = new_node
      else:
          new_node.next = self.head
          self.head.prev = new_node
          self.head = new_node

  def insert_at_end(self, data):
      new_node = DoublyNode(data)
      if self.head is None:
          self.head = new_node
          self.tail = new_node
      else:
          self.tail.next = new_node
          new_node.prev = self.tail
          self.tail = new_node
  def remove(self, key):
    if self.head is None:
      print("Lista vacia")
      return

    current= self.head
    while current is not None and current.data != key:
      current = current.next

    if current is None:
      print("Elemento {key} no encontrado")
      return

    if current == self.head:
      self.head = self.head.next
      if self.head:
        self.head.prev = None
      else:
        self.tail = None
    elif current == self.tail:
      self.tail = self.tail.prev
      self.tail.next = None
    else:
      current.prev.next = current.next
      current.next.prev = current.prev
  def search(self, target):

    current = self.head
    while current is not None:
      if current.data == target:
        return current
      current = current.next
    return None
  def traverse_forward(self):
    current = self.head
    while current is not None:
      print(current.data, end=" <-> ")
      current = current.next
    print("None")
  def traverse_backward(self):
    current = self.tail
    while current is not None:
      print(current.data, end=" <-> ")
      current = current.prev
    print("None")

  def mostrar(self):
      current = self.head
      while current:
          print(current.data, end=" <-> ")
          current = current.next
      print("None")

  def visualize(self):
      dot = Digraph(comment='Doubly Linked List')

      current = self.head
      while current:
          dot.node(str(current.data))
          if current.prev:
              dot.edge(str(current.prev.data), str(current.data), dir = "both")
          current = current.next

      display(dot)

lista = DoublyLinkedList()
lista.insert_at_end("Tarea1")
lista.insert_at_end("Tarea2")
lista.insert_at_end("Tarea3")
lista.insert_at_end("Tarea4")
lista.remove("Tarea2")
lista.visualize()
```
