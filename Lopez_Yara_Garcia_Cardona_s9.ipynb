```python
class Cola:
  def __init__(self):#Inicializa items
    self.items = []

  def encolar(self, item): # Encola los datos en la cola
    self.items.append(item)
    print("Encolado", item)

  def desencolar(self): # Desencola los datos en la cola
    if not self.esta_vacia(): #Revisa si la cola no está vacía
      item = self.items.pop(0)
      print("Desencolado", item)
    else:
      print("La cola está vacía")

  def esta_vacia(self): #Revisa si la cola está vacía
    return len(self.items) == 0

  def frente(self): #Obtiene el frente de la cola
    if not self.esta_vacia(): #Revisa si la cola no está vacía
      return self.items[0] #Devuelve el elemento en la posición 0, osea, el inicio de la cola
    else:
      print("La cola está vacía")
      return None

  def tamano(self): #Obtiene el tamaño de la cola
      return len(self.items) #Deuvelve la longitud de la lista

#Ejemplo de uso
if __name__ == "__main__":
  cola= Cola()
  cola.encolar("A")
  cola.encolar("B")
  cola.encolar("C")
  cola.desencolar()
  print(cola.frente())
  print(cola.tamano())
  print(cola.esta_vacia())
```

```python
class ColaEficiente:
  def __init__(self): # Se crean dos listas, una para encolar y la otra para desencolar
    self.itemsUsados = []
    self.itemsEliminados=[]

  def encolar(self, item): # Encolar los datos en la cola
    self.itemsUsados.append(item)
    print("Encolado", item)

  def desencolar(self): #desencolar los datos en la cola
    if self.eliminados_vacios(): #Revisa si la lista de eliminados está vacía
      item = self.itemsUsados.pop() #Obtiene el último valor de la lista y lo elimina de la cola
      self.itemsEliminados.append(item) #Guarda el valor en la lista de items usados

      print("Desencolado", item)
    else:
      while not self.esta_vacia(): #Mientras la lista de usados no esté vacía, elimina todos los datos y los guarda en la de eliminados
        item = self.itemsUsados.pop(0)
        self.itemsEliminados.append(item)
        print("Desencolado", item)


  def esta_vacia(self): #Revisa si la lista de usados está vacía
    return len(self.itemsUsados) == 0

  def eliminados_vacios(self): #Revisa si la lista de eliminados está vacía
    return len(self.itemsEliminados) == 0

  def frente(self): #Revisa el frente de la cola
    if not self.esta_vacia():
      return self.itemsUsados[0]
    else:
      print("La cola está vacía")
      return None

  def tamano(self): #Revisa el tamaño de la lista
      return len(self.itemsUsados)

  def visualize(self): #Nos permite visualizar las listas
    print(self.itemsUsados)
    print(self.itemsEliminados)

#Ejemplo de uso
if __name__ == "__main__":
  cola= ColaEficiente()
  cola.encolar("A")
  cola.encolar("B")
  cola.encolar("C")
  print(cola.visualize())
  cola.desencolar()
  print(cola.visualize())
  print(cola.frente())
  print(cola.tamano())
  print(cola.esta_vacia())
  cola.desencolar()
  print(cola.visualize())
  print(cola.frente())
  print(cola.tamano())
  print(cola.esta_vacia())
```

```python
class Cliente:
  def __init__(self, nombre, ticket): #Crea el objeto cliente que tenga nombre y ticket
    self.nombre = nombre
    self.ticket = ticket

  def __str__(self): #Permite visualizar el objeto
    return f"{self.nombre}"

class Cola:
  def __init__(self): #Crea el objeto cola
    self.items = []

  def encolar(self, item): #Encola los datos en la cola
    self.items.append(item)
    print("Encolado", item)

  def desencolar(self): #Desencola los datos en la cola
    if not self.esta_vacia():
      item = self.items.pop(0)
      print(item, "Atendido")
    else:
      print("La cola está vacía")

  def esta_vacia(self): #Revisa si la cola está vacía
    return len(self.items) == 0

  def frente(self): #Obtiene el frente de la cola
    if not self.esta_vacia():
      return self.items[0]
    else:
      print("La cola está vacía")
      return None

  def tamano(self): #Obtiene el tamaño de la cola
      return len(self.items)

  def visualize(self): #Visualiza la cola
    print([str(cliente) for cliente in self.items])

#Ejemplo de uso
if __name__ == "__main__":
  num_clientes = int(input("Ingrese el número de clientes: "))
  cola = Cola()
  for i in range(num_clientes):
    nombre=input("Ingrese su nombre: ")
    ticket= i+1
    cliente= Cliente(nombre, ticket)
    cola.encolar(cliente)

  print(cola.frente())
  print(cola.tamano())
  print(cola.esta_vacia())
  cola.visualize()
  cola.desencolar()
  print(cola.frente())
  print(cola.tamano())
  print(cola.esta_vacia())
  cola.visualize()
  cola.desencolar()
  print(cola.frente())
  print(cola.tamano())
  print(cola.esta_vacia())
  cola.visualize()
```

```python
class ColaCircular:
  def __init__(self, capacidad): #Crea una cola circular, que detecte el frente, el último valor, la capacidad y el tamaño de la cola
    self.items = [None] * capacidad
    self.capacidad = capacidad
    self.primero = 0
    self.ultimo = -1
    self.tamano = 0

  def encolar(self, item): #Encola los datos en la cola
    if self.tamano == self.capacidad: #Revisa si el tamaño es igual a la capacidad de la cola
      print("La cola está llena")
    elif self.esta_vacia(): #Revisa si la cola está vacía y ejecuta este código
      self.primero = 0 #El frente será 0
      self.ultimo = 0 #El último dato de la cola está en posición 0
      self.items[self.ultimo] = item #Agrega el item a la cola
      self.tamano += 1 #Agrega 1 al tamaño
      print("Encolado", item)
    else: #Si ya hay valores en la cola, ejecuta
      self.ultimo = (self.ultimo + 1) % self.capacidad #El último valor será el residuo del anterior último valor + 1 con la capacidad de la cola
      self.items[self.ultimo] = item #guarda en el último valor el item
      self.tamano += 1 #Agrega 1 al tamaño
      print("Encolado", item)

  def desencolar(self): #Desnecola los datos de la lista
    if self.esta_vacia(): #Revisa si la cola está vacía
      print("La cola está vacía")
    elif self.primero == self.ultimo: #Revisa si solo hay un dato en la cola
      item = self.items[self.primero] #Obtiene el item de la cola
      self.items.pop(0) #Elimina el item
      self.primero = 0 #El frente de la cola es 0
      self.ultimo = -1 #El último valor será -1, debido a que ya no hay datos
      self.tamano -= 1 #Resta 1 al tamaño
      print("Desencolado", item)
    elif self.primero < self.capacidad - 1: #Revisa si el frente es menor a la capacidad -1
      item = self.items[self.primero] #Obtiene el item de la cola
      self.items.pop(0) #Elimina el item
      self.primero = 0 #El frente de la cola es 0
      self.tamano -= 1 #Resta el tamaño de la cola
      print("Desencolado", item)

  def esta_vacia(self): #Revisa si la cola está vacía
    return self.tamano == 0

  def frente(self): #Revisa el frente de la lista
    if not self.esta_vacia():
      return self.items[self.primero]
    else:
      print("La cola está vacía")
      return None

  def tamaño(self): #Obtiene el tamaño de la lista
    return self.tamano

  def visualize(self): #Visualiza la lista
    print(self.items)


#Ejemplo de uso
if __name__ == "__main__":
  cola= ColaCircular(15)
  for i in range(15):
    cola.encolar(i+1)
  cola.visualize()
  cola.desencolar()
  cola.visualize()
  print(cola.frente())
  print(cola.tamaño())
  print(cola.esta_vacia())
  cola.desencolar()
  cola.visualize()
  print(cola.frente())
  print(cola.tamaño())
  print(cola.esta_vacia())
```
